# Lab 01 Terminology

Key FastAPI concepts you should understand before starting the Lab 01 exercises.


## Scaffolding

**Definition:** Creating the starter project structure and files generated by the lab scripts so you can focus on the API logic instead of manual setup.

**In Lab 01:** In Lab 01 you run the provided project template to scaffold routes, settings, and directories that you will immediately customize.

## Endpoints

**Definition:** The individual URL routes that accept requests and return responses, typically implemented as FastAPI path operation functions.

**In Lab 01:** You add multiple endpoints to expose lab data and connect them to the React client once the FastAPI app is scaffolded.

## Validation

**Definition:** Automatic checking of incoming request data against declared types and constraints so that only valid payloads reach your business logic.

**In Lab 01:** FastAPI validates incoming JSON using Pydantic models you define while implementing Lab 01 routes.

## Serialization

**Definition:** Transforming Python objects into JSON responses (and vice versa) according to the schemas you describe.

**In Lab 01:** Your Lab 01 endpoints return Pydantic models which FastAPI serializes into JSON for the React CRA front end.

## Asynchronous Processing

**Definition:** Writing non-blocking `async` endpoint functions that can handle multiple concurrent requests efficiently.

**In Lab 01:** Lab 01 introduces async route handlers so your API can serve the React client without waiting on slow operations.

## Dependency Injection

**Definition:** FastAPI's way of providing shared resources—such as configuration, database sessions, or clients—to routes through callable dependencies.

**In Lab 01:** While wiring your Lab 01 app, you register dependencies that make shared services available to the endpoints you scaffold.

## OpenAPI Documentation

**Definition:** An automatically generated API specification and interactive docs that describe your routes, parameters, and responses.

**In Lab 01:** After running the Lab 01 server you inspect the `/docs` page to confirm the endpoints you added are documented correctly.
